```python
import pandas as pd
```

## Data visualization with pandas

```python
df = pd.DataFrame({"Student":["Aman","Ram","Shyam","Kana","Ghanu"],"Maths":[85,98,90,60,78],"Science":[78,55,91,65,82],"English":[90,2,87,70,85],
                   "Hours_studied":[4,22,6,2,5],"Attendance":[95,54,98,80,92],"Gender":["M",'F','M','F','M']})
df
```

```python
# Line plot
df.plot()
```

```python
# Area plot
df.plot.area(alpha=0.8)
```

```python
df.plot.bar()
```

```python
df.plot.hist(bins=50)
```

```python
# Scatter plot --> Passing x and y label is necessary
df.plot.scatter(x='Science',y='Maths')
```

```python
# Box plot
df.plot.box()
```

```python
df.plot.hexbin(x = 'Maths', y="Science", gridsize=15, cmap = "Blues")
```

```python
df["English"].plot.kde()
```
