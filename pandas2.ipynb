### Boolean indexing in pandas 

```python
# dictionary of lists
dict = {'name':["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score':[90, 40, 80, 98]}

# creating a dataframe with boolean index 
dfd = pd.DataFrame(dict, index = [True, False, True, False]) # it creates the boolean index for each row
dfd
# accessing a dataframe using .loc[] function 
#print(dfd.loc[True])  # Selecting only those row having boolean index true
```

### DataFrame.truncate()

```python
# To remove rows and columns from the dataset we use dataframe.truncate()
# Examples --> Creating dataframe 
Y = {"Name":["Ram","Raju","Ravi","Dharmraj","Shyam"],"Class":[10,11,10,12,10],"Stream":["Science","PCM","Science","PCM","Science"],"Roll_no":[101,102,103,104,105],"strem_id":["S-18","M-25","S-18","M-25","S-18"]}

# Creating index using date (New concept)
index_date = pd.date_range('2010-10-09 08:45', periods = 5, freq ='YE') # Here we can change the frequncey parameter to increase the index value 
dfy = pd.DataFrame(Y,index = index_date)                                # Here we can pass (hour = h, month = ME, Year = YE, etc)
dfy
```

```python
# Now using truncate() function to remove rows (truncate() remove rows by default)
row_truncate = dfy.truncate(before = "2011-12-31 08:45:00", after = "2013-12-31 08:45:00")
row_truncate # This result contain rows before "2011-12-31 08:45:00" and after "2013-12-31 08:45:00" 
```

```python
# Truncate by column (Remember to pass the before, after parameter in sorted order)
dfy_sorted = dfy.sort_index(axis = 1) # axis = 1 means sorted the column name 
dfy_sorted
col_truncate = dfy_sorted.truncate(after="Stream",before="Name",axis = 1)
col_truncate
```

```python
# We can truncate the row and column using iloc function also
dfy.iloc[:,1:4]  # The first colon indicate that we are including aLL the rows , after comma we are selecting columns
dfy.iloc[1:3,1:4]
```

### Iterating over the rows and column of the dataframe

```python
# Iterating over rows --> using itertuples()
for row in df.head(2).itertuples():
    print(row)
    print(" ")
```

```python
# Iterating over column --> using items()
for label,val in df.head(5).items():
    print(f"Columns: {label}")
    print(val)
    print()
```

## Adding new rows to the existing data frame

```python
data = {"Name" : ["Ram","Shyam"],"Age":[20,21]}
df = pd.DataFrame(data)
df
# Add row using loc function
df.loc[len(df)] = ["Charlie",25] # We have to specify the index where we add the row 
df
# Add row using concate function 
new_row = pd.DataFrame({"Name":["Bob"],"Age":[24]})
df = pd.concat([df,new_row],ignore_index = True)
df
# Adding multiple rows 
rows = pd.DataFrame({"Name":["Dheeraj","Prem","winston"],"Age":[20,26,28]})
df = pd.concat([df,rows],ignore_index = True)
df
# Adding row without one column
row = {"Name":"Alice"}
df = pd.concat([df,pd.DataFrame([row])],ignore_index = True)
df
```
