```python
import pandas as pd
```

## Grouping data in pandas

```python
# Most common group by operations --> 
# mean(), sum(), count(), size(), max(), min(), median(), std(), agg()
```

```python
df = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "HR", "Sales", "IT"],
    "Salary": [50000, 40000, 60000, 45000, 55000, 70000],
    "Age": [25, 30, 28, 30, 27, 25]
})
df
df.groupby("Department") # It returns an object over which we can apply functions
df.groupby("Department")["Salary"].sum()
df.groupby("Department")["Salary"].mean()
df.groupby("Department",as_index = False).count()
```

```python
# Applying multiple aggregate functions simultaneoulsy
df.groupby("Department",as_index = False)["Salary"].agg(["mean", "max", "min"])
```

```python
# Grouping by multiple columns
df.groupby(["Department","Age"],as_index = False)["Salary"].sum()
```

## Iterating over groups

```python
grp = df.groupby("Department")
for name,group in grp:
    print(name)  # This print the name of the group
    print(group)  # This print the data inside the group 
    print()
```

## Selecting a particular group from groups

```python
grp1 = df.groupby("Department")
grp1.get_group("IT") # It return the group of IT sector
```

## Combining multiple column using dictionary aggregate function 

```python
data = pd.DataFrame({
    'Store': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Product': ['Apple', 'Banana', 'Apple', 'Banana', 'Apple', 'Banana'],
    'Sales': [100, 150, 200, 100, 120, 180],
    'Quantity': [10, 20, 30, 40, 15, 35]
})
agg_dict = {"Sales":'sum',"Quantity":'mean'}  # Sum for Sales and mean for quantity
result = data.groupby("Store").agg(agg_dict).reset_index()
result
data
```

```python
# Multiple aggregation on single column
agg_dict1 = {"Sales":['max','min'],"Quantity":['max','min','median']}
r = data.groupby("Store").agg(agg_dict1).reset_index()
r
```

```python
# We can create our own custom aggregate function by using lambda functions
# Similarly we can create any kind of function according to our need 
agg_custom = {"Sales": lambda x: x.max()-x.min(), 'Quantity':'sum'}

re = data.groupby("Store").agg(agg_custom).reset_index()
re
```

## Joining two dataset of same nature

```python
# Creating custom dataset
data1 = pd.DataFrame({'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']},index=[0,1,2,3])

data2 = pd.DataFrame({'Name': ['Abhi', 'Ayushi', 'Dheeraj', 'Hitesh'],
         'Age': [17, 14, 12, 52],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']},index=[4,5,6,7])
# Combining using pd.concat function
# After combining data1 and data2 
data1_data2_list = [data1,data2]
result = pd.concat(data1_data2_list)
result
```
