```python
import pandas as pd 
import numpy as np
```

## Pandas series 

```python
# Pandas series --> 1-D array,creation 
s = pd.Series()
print(s)
data = np.array([1,2,3,4,5,8,9,12,54])
ser = pd.Series(data*2) # Using numpy broadcasting 
print(ser)
#Indexing -> 1 position based indexing 
print(ser[:5]) # Return first 5 values

#Indexing -> 2. Label based index --> Same as python indexing 
print(ser[3])
```

```python
# Different ways to create pandas series 
# 1. Using dictionary 
dict_data = {'Geeks': 10, 'for': 20, 'geeks': 30} # The keys of the dictionary becomes the index of the series 
ser = pd.Series(dict_data)
print(ser)

# 2. Using range function 
data = pd.Series(range(1,10,2))
print(data)

# 3. Using numpy function
ser1 = pd.Series(np.linspace(1, 10,4)) # Creates equally space 4 numbers between 1 and 10
print(ser1)

# 4. Using list comprehension
ser2 = pd.Series([i for i in "abcdefg"])
print(ser2)
```

```python
# In indexing we can provide our own index while creating the pandas series 
data = ['a','s','r','o','g','r','t']
ser = pd.Series(data,index = [10,11,12,13,14,15,16]) # We can provide the index parameter 
print(ser[13])
```

```python
# Reading csv files and making pandas series from the IPL data set.
df = pd.read_csv("matches.csv")
#df.head(10)
data = pd.Series(df["city"]) # The entire city column is now available in data 
data.loc[2:10]  # loc function uses the position based indexing that's why it includes both the start and end position
data.iloc[2:10]  # iloc funtion uses integer based indexing that is why it exclude the end position
```

```python
# Binary operation on pandas series --> The indexing of the series should be same 
ser1 = pd.Series([1.0, 2.0, 3.0], index=['A', 'B', 'C'])
ser2 = pd.Series([4, 5, 6], index=['A', 'B', 'C'])
# We have multiple binary operation like --> add(),sub(),mul(),mean(),pow() etc
df_sum = ser1.add(ser2)
print(df_sum)
df_sub = ser1.sub(ser2) # df_sub = ser1 - ser2
print(df_sub)
```

```python
# making a dict of list ---> Just an example to understant the sum() function for the binary operation 
data = {'name': ['John', 'Peter', 'Karl'],
        'age' : [23, 42, 19],
        'salary':[15000,20000,18000]}
val = pd.DataFrame(data)
# sum of all salary --> Binary operation on age column
val['total_salary'] = val['salary'].sum() # Adding new column to the already existing data set
val["e_id"] = ['a','b','c']
val
```

```python
# Conversion of data type in pandas 
ser = pd.Series([1,2,3,4])
ser = ser.astype(float) # converting integer to float
for i in ser:   # we can apply loops,condtions on the pandas series
    if(i==3.0):
        continue  
    print(i)
    
```

## Pandas Dataframe

```python
# Creating dataframe  --> There are multiple ways to create the DataFrame in pandas. 
# 1. Creating an empty dataframe
df = pd.DataFrame()
print(df)
print("*"*20)

# 2. Creating dataframe from a list
sports = ['Cricket', 'Football', 'Volleyball', 'Badminton', 'Basketball', 'Kho-Kho', 'Table Tenis']
df1 = pd.DataFrame(sports)
print(df1)
print("*"*20)

# 3. Creating dataframe from numpy array and dictionary
data = {'A':np.array([1,2,3]),'B':np.array([4,5,6]),'C':np.array([7,8,9])}
df2 = pd.DataFrame(data)
print(df2)
print("*"*20)

# 4. Creating dataframe from list of dictionary
data = [
    {'Name': 'Mike', 'Degree': 'MBA', 'Score': 90},
    {'Name': 'Dan', 'Degree': 'BCA', 'Score': 40},
    {'Name': 'Emilia', 'Degree': 'M.Tech', 'Score': 80},
]
df3 = pd.DataFrame(data)
print(df3)
print("*"*20)
# 5. Creating dataframe from a dictionary of list
info = {'name': ['John', 'Peter', 'Karl'],
        'age' : [23, 42, 19],
        'salary':[15000,20000,18000]}
val = pd.DataFrame(info)
#print(val)
val
```
